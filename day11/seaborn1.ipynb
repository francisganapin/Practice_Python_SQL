```python
import pandas as pd

data = {
    'Name': ['Anna', 'Ben', 'Charlie', 'Daisy', 'Ella', 'Frank', 'Gina', 'Henry', 'Ivy', 'Jack'],
    'Age': [23, 30, 31, 29, 22, 35, 27, 40, 25, 28],
    'Gender': ['F', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F', 'M'],
    'BMI': [21.5, 25.2, 27.8, 23.1, 20.7, 29.3, 22.0, 26.7, 19.8, 24.4],
    'WorkoutHoursPerWeek': [5, 3, 4, 6, 7, 2, 6, 3, 8, 4],
    'City': ['Manila', 'Cebu', 'Davao', 'Cebu', 'Manila', 'Davao', 'Manila', 'Cebu', 'Manila', 'Davao']
}

df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df,x='BMI',kde=True,bins=8)
```

```python
sns.scatterplot(data=df,x='BMI',y='WorkoutHoursPerWeek',hue='Gender')
plt.title('BMI vs Workout Hours')
plt.show()
```

```python
sns.boxplot(data=df,x='Gender',y='BMI')
plt.title('BMI by Gender')
plt.show()
```

```python
sns.barplot(data=df,x='City',y='WorkoutHoursPerWeek')
plt.title('Average Workout Hours by City')
plt.show()
```

```python
corr = df[['Age','BMI','WorkoutHoursPerWeek']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
