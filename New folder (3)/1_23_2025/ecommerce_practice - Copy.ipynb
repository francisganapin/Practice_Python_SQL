## Question 3: Identify Refunded Transactions

**Find all orders that have been refunded, including the customer name, order date, and refund details.**

Use JOIN to connect orders, customers, and refunds tables.

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    o.order_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    o.order_date,
    o.total_amount AS original_amount,
    r.refund_date,
    r.refund_amount,
    r.reason
FROM refunds r
JOIN orders o ON r.order_id = o.order_id
JOIN customers c ON o.customer_id = c.customer_id
ORDER BY r.refund_date;
"""
run_query(query)
```

## Question 4: Customer Transaction History

**Create a complete transaction history for customer "Emily Johnson" (customer_id = 2), showing all orders with their items.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    c.first_name || ' ' || c.last_name AS customer_name,
    o.order_id,
    o.order_date,
    o.status,
    p.product_name,
    oi.quantity,
    oi.subtotal
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
WHERE c.customer_id = 2
ORDER BY o.order_date, o.order_id;
"""
run_query(query)
```

## Question 5: Payment Method Analysis

**Analyze transactions by payment method. Show the total revenue and number of transactions for each payment method.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    payment_method,
    COUNT(*) AS transaction_count,
    SUM(amount) AS total_revenue,
    ROUND(AVG(amount), 2) AS avg_transaction
FROM payments
WHERE status = 'completed'
GROUP BY payment_method
ORDER BY total_revenue DESC;
"""
run_query(query)
```

## Question 6: Cancelled Orders Analysis

**Find all cancelled orders and calculate the total revenue lost from cancellations. Include what products were in those orders.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    o.order_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    o.order_date,
    p.product_name,
    oi.quantity,
    oi.subtotal,
    o.total_amount AS order_total
FROM orders o
JOIN customers c ON o.customer_id = c.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
WHERE o.status = 'cancelled';
"""
print("Cancelled Orders Details:")
display(run_query(query))

# Total lost revenue
query2 = """
SELECT 
    COUNT(DISTINCT order_id) AS cancelled_orders,
    SUM(total_amount) AS total_lost_revenue
FROM orders
WHERE status = 'cancelled';
"""
print("\nTotal Revenue Lost:")
display(run_query(query2))
```

## Question 7: Top Spending Customers

**Find the top 5 customers by total spending (completed orders only). Include their email and total orders count.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    c.email,
    COUNT(o.order_id) AS total_orders,
    SUM(o.total_amount) AS total_spent
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
WHERE o.status = 'completed'
GROUP BY c.customer_id
ORDER BY total_spent DESC
LIMIT 5;
"""
run_query(query)
```

## Question 8: Product Sales Tracking

**Find the best-selling products by quantity sold. Include the category name and total revenue generated by each product.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    p.product_name,
    cat.category_name,
    SUM(oi.quantity) AS total_quantity_sold,
    SUM(oi.subtotal) AS total_revenue
FROM products p
JOIN categories cat ON p.category_id = cat.category_id
JOIN order_items oi ON p.product_id = oi.product_id
JOIN orders o ON oi.order_id = o.order_id
WHERE o.status NOT IN ('cancelled')
GROUP BY p.product_id
ORDER BY total_quantity_sold DESC;
"""
run_query(query)
```

## Question 9: Order Status Summary

**Create a summary showing the count and total value of orders for each status (pending, shipped, completed, refunded, cancelled).**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    status,
    COUNT(*) AS order_count,
    SUM(total_amount) AS total_value,
    ROUND(AVG(total_amount), 2) AS avg_order_value
FROM orders
GROUP BY status
ORDER BY order_count DESC;
"""
run_query(query)
```

## Question 10: Monthly Revenue Trend

**Analyze the monthly revenue trend. Group completed orders by month and show the revenue progression.**

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    strftime('%Y-%m', order_date) AS month,
    COUNT(*) AS order_count,
    SUM(total_amount) AS monthly_revenue,
    ROUND(AVG(total_amount), 2) AS avg_order_value
FROM orders
WHERE status = 'completed'
GROUP BY strftime('%Y-%m', order_date)
ORDER BY month;
"""
run_query(query)
```

## Question 11: Full Transaction Backtrack

**For a complete audit, create a query that shows the full lifecycle of order ID 7 (a refunded order):**
- Customer info
- Order details
- Items purchased
- Payment info
- Refund info

```python
# YOUR CODE HERE

```

```python
# SOLUTION
# Part 1: Order and Customer Info
query1 = """
SELECT 
    o.order_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    c.email,
    o.order_date,
    o.status,
    o.total_amount,
    o.shipping_address
FROM orders o
JOIN customers c ON o.customer_id = c.customer_id
WHERE o.order_id = 7;
"""
print("üì¶ ORDER INFO:")
display(run_query(query1))

# Part 2: Items in the order
query2 = """
SELECT 
    p.product_name,
    oi.quantity,
    oi.unit_price,
    oi.subtotal
FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
WHERE oi.order_id = 7;
"""
print("\nüõçÔ∏è ORDER ITEMS:")
display(run_query(query2))

# Part 3: Payment info
query3 = """
SELECT 
    payment_id,
    payment_date,
    amount,
    payment_method,
    status
FROM payments
WHERE order_id = 7;
"""
print("\nüí≥ PAYMENT INFO:")
display(run_query(query3))

# Part 4: Refund info
query4 = """
SELECT 
    refund_id,
    refund_date,
    refund_amount,
    reason,
    status
FROM refunds
WHERE order_id = 7;
"""
print("\nüîô REFUND INFO:")
display(run_query(query4))
```

## Question 12: Category Performance Analysis

**Analyze the performance of each product category:**
- Total products sold
- Total revenue
- Number of orders
- Average order value per category
- Refund rate (percentage of refunded orders)

```python
# YOUR CODE HERE

```

```python
# SOLUTION
query = """
SELECT 
    cat.category_name,
    COUNT(DISTINCT o.order_id) AS total_orders,
    SUM(oi.quantity) AS total_items_sold,
    ROUND(SUM(oi.subtotal), 2) AS total_revenue,
    ROUND(AVG(oi.subtotal), 2) AS avg_item_value,
    SUM(CASE WHEN o.status = 'refunded' THEN 1 ELSE 0 END) AS refunded_orders,
    ROUND(
        SUM(CASE WHEN o.status = 'refunded' THEN 1.0 ELSE 0 END) / COUNT(DISTINCT o.order_id) * 100, 
        2
    ) AS refund_rate_percent
FROM categories cat
JOIN products p ON cat.category_id = p.category_id
JOIN order_items oi ON p.product_id = oi.product_id
JOIN orders o ON oi.order_id = o.order_id
GROUP BY cat.category_id
ORDER BY total_revenue DESC;
"""
run_query(query)
```

# üéØ Bonus Challenge

Create your own query to:
1. Find customers who have never made a purchase
2. Identify products that have never been sold
3. Calculate the customer lifetime value (CLV)

```python
# BONUS: Customers who never made a purchase
query = """
SELECT 
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    c.email,
    c.registration_date
FROM customers c
LEFT JOIN orders o ON c.customer_id = o.customer_id
WHERE o.order_id IS NULL;
"""
print("Customers with no purchases:")
display(run_query(query))
```

```python
# BONUS: Products never sold
query = """
SELECT 
    p.product_id,
    p.product_name,
    cat.category_name,
    p.price,
    p.stock_quantity
FROM products p
JOIN categories cat ON p.category_id = cat.category_id
LEFT JOIN order_items oi ON p.product_id = oi.product_id
WHERE oi.item_id IS NULL;
"""
print("Products never sold:")
display(run_query(query))
```
