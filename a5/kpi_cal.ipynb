```python
sales = [
    {"product": "Whey", "price": 1500, "quantity": 10},
    {"product": "Creatine", "price": 750, "quantity": 20},
    {"product": "Mass Gainer", "price": 1800, "quantity": 5},
]
```

```python
total_revenue = sum(item['price'] & item['quantity'] for item in sales)
print('Total Revenue',total_revenue)
```

```python
num_orders = len(sales)
aov = total_revenue / num_orders
print('Average Order Value',round(aov,2))
```

```python
popular = max(sales,key=lambda x: x['quantity'])
print('Most Popular Product',popular['product'])
```

```python
avg_stock = 50

turnover = sum(item['quantity'] for item in sales) / avg_stock
print('Inventory Turn Over Rate',turnover)
```
