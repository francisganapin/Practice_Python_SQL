```python
my_array = [64,34,25,25,12,22,11,90,5]

n = len(my_array)
for i in range(n-1):
    for j in range(n-i-1):
        if my_array[j] > my_array[j+1]:
            my_array[j],my_array[j+1] = my_array[j+1],my_array[j]


print('Sorted array',my_array)
```

```python
data = [64,34,25,12,22,11,90]
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1],arr[j]
    return arr


print('Bubble sorted:',bubble_sort(data.copy()))
```

```python
print('Built-in Sorted',sorted(data))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'product':['Whey Protein','Creatine','Mass Gainer','Pre-Workout'],
    'revenue':[15000,12000,10000,8000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['revenue'].plot(kind='bar',color='skyblue')
```

```python
df['revenue'].plot(kind='bar',color='skyblue')
plt.xticks(ticks=range(len(df)),labels=df['product'],rotation=45)
plt.xlabel('Products')
plt.ylabel('Revenue ($)')
plt.title('Supplement Sales Revenue')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'product':['Whey','Creatine','Mass Gainer','Pre-Workout','Whey','Creatine'],
    'sales':[200,150,100,80,220,180],
    'month':['Jan','Jan','Jan','Jan','Feb','Feb']
}

df = pd.DataFrame(data)
```

```python
sns.barplot(x="product", y="sales", data=df, estimator=sum, ci=None, palette="Blues")
plt.title('Total Sales By Product')
plt.show()
```
