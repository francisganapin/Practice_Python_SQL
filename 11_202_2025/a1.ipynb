```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data_100.csv')
```

```python
print('Data preview')
print(df.head())
```

```python
city_sales = df.groupby('city')['amount'].sum().reset_index()
print('Total sales per City')
print(city_sales)
```

```python
product_sales = df.groupby('product')['amount'].sum().reset_index()
top_product = product_sales.loc[product_sales['amount'].idxmax()]
print('Product with highest revenue')
print(top_product)
```

```python
avg_qty = df['quantity'].mean()
print(f'Average quantity per transaction:{avg_qty}')
```

```python
import matplotlib.pyplot as plt
df['date'] = pd.to_datetime(df['date'])
daily = df.groupby('date')['amount'].sum()
daily.plot(kind='line')
plt.title('Daily Sales')
plt.ylabel('Amount')
plt.show()
```

```python
df2 = pd.read_csv('alt_data_100.csv')


region_rev = df2.groupby('region')['revenue'].sum()
print(region_rev)
```

```python
cat_rev = df2.groupby('product_category')['revenue'].sum().reset_index()
top_cat = cat_rev.loc[cat_rev['revenue'].idxmax()]
print(top_cat)
```

```python
ads = pd.read_csv('ads_spend_100.csv')


spend_by_data = ads.groupby('channel')['spend'].sum()
print('Total Spend per channel:',spend_by_data)
```

```python
roi = ads.groupby('channel').apply(lambda  d: d['revenue'].sum() / d['spend'].sum())
print(roi)
```

```python
#daily Spend trend (Line chart)
ads['date'] = pd.to_datetime(ads['date'],errors='coerce')
daily = ads.groupby('date')['spend'].sum()
daily.plot(kind='line',title='Daily Spend Trend Ads')
plt.ylabel('Spend')
plt.show()
```

```python
total_spend  = ads['spend'].sum()
print(total_spend)
```
