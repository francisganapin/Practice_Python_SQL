```python
class Car:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

    def drive(self):
        print(f"{self.brand} {self.model} is driving")
```

```python
class ElectricCar(Car):
    def drive(self):
        print(f"{self.brand} {self.model} is driving electrically")
```

```python
car1 = Car("Toyota","Corolla")
car2 = ElectricCar("Tesla","Model 3")

car1.drive()
car2.drive()
```

```python
from abc import ABC,abstractmethod


class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Aso"

class Cat(Animal):
    def sound(self):
        return "Pusa"


animals = [Dog(),Cat()]
for a in animals:
    print(a.sound())
```
