```python
words = ["eat","tea","tan","ate","nat","bat"]

group = {}

for w in words:
    key = "".join(sorted(w))
    if key not in group:
        group[key] = []
    group[key].append(w)

print(list(group.values()))
```

```python
def count_digits_manual(n):
    s = str(n)

    counts = {}
    for digit in s:
        if digit.isdigit():
            counts[digit] = counts.get(digit,0) + 1

    return counts


numbers = "112233145"
print(count_digits_manual(numbers))
```
