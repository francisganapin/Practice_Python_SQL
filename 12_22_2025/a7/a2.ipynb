```python
def  count_digits_manual(n):
    s = str(abs(n))

    counts = {}
    for digit in s:
        if digit.isdigit():
            counts[digit] = counts.get(digit,0) + 1
    return counts

numbers = "112233145"
print(count_digits_manual(numbers))
```

```python
def count_digits_manual(n):
    s = str(n)

    counts = {}
    for digit in s:
        if digit.isdigit():
            counts[digit] = counts.get(digit,0) + 1

    return counts


numbers = "112233145"
print(count_digits_manual(numbers))
```

```python
def http_error(status):
    match status:
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 418:
            return "I'm a teapot"
        case _:
            return "Something's wrong with the internet"
        

print(http_error(400))
```
